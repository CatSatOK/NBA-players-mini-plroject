## Analyse des données

Regarder l'évolution des différentes informations fournies pour chaque joueur et comparer les afin d'en savoir plus sur ce qui peut influencer les salaires.

Pour cela, commencer par construire une fonction __*courbes(data,joueurs)*__ :
- qui prend en argument :
>- *data* : les données que l'on veut afficher (par exemple Salary ou encore FreeThrows/FreeThrowsAttempts pour le taux de réussite aux lancer-francs ou ce qu'on veut)
>- *joueurs* : la liste de joueurs dont on veut afficher les informations
- qui retourne :
>- une figure avec l'ensemble des courbes de chaque joueur sur le même graphique

Cette fonction vous permettra d'avoir une première analyse des données et de leur évolution en affichant les données sous forme de courbes.

### Des pistes pour l'analyse

- Des problématiques auxquelles répondre :
>- évolution des salaires

>- évolution et impact des temps de jeu
>- les styles de jeu :
>>- précision des joueurs

>>- impact des lancer-francs

```python
def playerinfo(player, season):
    ind = Players.index(player)
    yr = Seasons.index(season)
    print(f"Salary for {player} in {season} was:$ {Salary[ind][yr]}")

playerinfo("LeBronJames", "2007")
```

```python
def playergraph(player):
    ind = Players.index(player)
    
    x = Seasons
    y = Salary[ind]
    plt.plot(x, y, label = "Salary")

    plt.xlabel('Season')
    plt.ylabel('Salary in millions')
    plt.title(f'Players wages over time: {player}')
    plt.show()

playergraph("KobeBryant")
```

```python
def precision(player):
    ind = Players.index(player)
    
    fig,ax = plt.subplots()
    ax.plot(Seasons, FieldGoalAttempts[ind], color="green", marker="o")
    ax.set_xlabel("Seasons",fontsize=14)
    ax.set_ylabel("Field Goal Attempts",color="green",fontsize=14)
    ax2=ax.twinx()

    ax2.plot(Seasons, Points[ind],color="blue",marker="o")
    ax2.set_ylabel("Points",color="blue",fontsize=14)
    plt.show()

precision("DwayneWade")
```

```python
def freethrow(player):
    ind = Players.index(player)
    
    x = Seasons
    y = FreeThrowsAttempts[ind]
    plt.plot(x, y, label = "Free Throw attempts")
    
    x2 = Seasons
    y2 = FreeThrows[ind]
    plt.plot(x2, y2, label = "Free Throws made")

    plt.xlabel('Seasons')
    plt.ylabel('Free Throws')
    plt.title(f'Free throw success : {player}')
    plt.legend()
    plt.show()

freethrow("DwayneWade")
```

```python
def wages_per_min(player, season):    
    ind = Players.index(player)
    yr = Seasons.index(season)
    pay_per_min = Salary[ind][yr] / MinutesPlayed[ind][yr]
    print(f"{player} wages per minute played in {season} = {pay_per_min}")
        
wages_per_min("KobeBryant", "2005")
```

```python
def wpm_graph(player):
    ind = Players.index(player)
    
    x = Seasons
    y = Salary[ind] / MinutesPlayed[ind]
    plt.plot(x, y, label = "Salary per min")

    plt.xlabel('Season')
    plt.ylabel('Salary per min')
    plt.title(f'Players wages per minute played: {player}')
    plt.show()
    
wpm_graph("LeBronJames")
```

```python
def points():
    for player in Players:
        ind = Players.index(player)
        x = Seasons
        y = Points[ind]
        plt.legend(bbox_to_anchor=(1.05, 1))
        plt.plot(x, y, label = player)
        plt.xlabel('Season')
        plt.ylabel('Points')
        plt.title('Points')
points()
```

```python
def salary():
    for player in Players:
        ind = Players.index(player)
        x = Seasons
        y = Salary[ind]
        plt.legend(bbox_to_anchor=(1.05, 1))
        plt.plot(x, y, label = player)
        plt.xlabel('Season')
        plt.ylabel('Salary in millions')
        plt.title('Salary')

salary()
```

```python
def everything(player, season):
    ind = Players.index(player)
    yr = Seasons.index(season)
    print(f"In {season} {player} was paid {Salary[ind][yr]}, played {Games[ind][yr]} games, played {MinutesPlayed[ind][yr]} minutes, made {FieldGoals[ind][yr]} field goals, had {FieldGoalAttempts[ind][yr]} field goal attempts, scored {Points[ind][yr]} points, had {FreeThrows[ind][yr]} free throws and {FreeThrowsAttempts[ind][yr]} free throw attempts")

everything("KevinDurant", "2009")    
    
```
